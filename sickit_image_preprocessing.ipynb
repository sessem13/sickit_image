```python
from matplotlib import pyplot as plt
```

```python
from skimage import io ## to read images
```

```python
img1 = io.imread('/content/poultry2.jpg')
print(img1)
print(img1.size)
print(img1.shape)
print(type(img1))
```

```python
plt.imshow(img1)
```

```python
from skimage import transform as tf ## for transformation functions usage:
```

```python
transform = tf.AffineTransform(scale=1, rotation = 0.9, translation=[50,100])
```

```python
transformed_image = tf.warp(img1, transform)
plt.imshow(transformed_image)
```
